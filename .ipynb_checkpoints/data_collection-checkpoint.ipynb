```python
from djitellopy import tello
from djitellopy import swarm
from djitellopy import enforce_types
import pandas as pd
import threading
import time
pd.set_option('display.max_columns', None)
pd.set_option("max_rows", None)
```

```python
#Create swarm based on ip address from txt file

s=swarm.TelloSwarm.fromFile("C:/Users/Kew/Desktop/project/drone_project/drone_datacollect/djitellopy/ips.txt")
```

```python
#define function to print out after 30 sec of hovering
def t_down():
    print("10 sec")
    s.parallel(lambda i, tello: tello.move_down(20))
def t_up():
    print("10 sec")
    s.parallel(lambda i, tello: tello.move_up(20))
def t_land():
    print("10 sec")
    s.parallel(lambda i, tello: tello.land())
t1 = threading.Timer(10,t_up)
t2 = threading.Timer(20,t_down)
t3 = threading.Timer(30.0, t_land)

#Put all drones in swarm in command mode
s.parallel(lambda i, tello: tello.connect())
#Tell all drones in swarm to take off
s.parallel(lambda i, tello: tello.takeoff())
#Tell all drones in swarm to go down by 30 cm to adjust to be in front of ventilator
s.parallel(lambda i, tello: tello.move_down(30))
#Hover 30 sec and land (the up and down are added as tello somehow has an auto land after 15 sec or so)
t1.start()
t2.start()
t3.start()

#Tell all drones in swarm to land
#s.parallel(lambda i, tello: tello.land())
```

```python
#Register data from swarm internal log
d=tello.STATE_LOG
#Tell all drones in swarm to end their instance
s.parallel(lambda i, tello: tello.end())
```
